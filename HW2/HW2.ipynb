Data pre-processing

```python

df = pd.read_csv('2014-2020.csv')
df["Total"] = df["AES"]+df["TES"]+df["VDE"]+df["TEC"]+df["GES"]+df["GAES_GEN"]
df['Hour'] = [date.split("-")[0] for date in df["TIME"]]
df['Date'] = pd.to_datetime([date.split("-")[1].replace(".","-") for date in df["TIME"]])
df
```

- Як змінювалась структура генерації електроенергії за роками?


```python
df["Year"] = df["Date"].dt.year
__df_years = df.groupby(["Year"], as_index=False).sum()
df_tall_year = pd.melt(__df_years, id_vars=['Year'], value_vars=['AES','TEC', 'VDE', 'TES',  'GES','GAES_GEN'], value_name="Energy",var_name="Type")
df_tall_year.head()
```

```python
# chart_year_area = alt.Chart(df_tall_year).mark_area(opacity = 0.5, ).encode(
#     y = alt.Y('Energy:Q', sort=alt.Sort(field='Energy', op='sum', order='ascending'), stack='zero'),
    
#     # x = alt.X('Type:N', sort=alt.Sort(field='Energy', op='sum', order='descending')),
#     x = alt.X('Year:N'),
#     color=alt.Color("Type:N")

#     # color=alt.Color("Year:N")

#     # row = alt.Row('Year:O'),
# )

chart_year_area = (alt.Chart(df_tall_year).mark_line(size = 2, interpolate='natural',point = True)
                    + alt.Chart(df_tall_year).mark_area(opacity = 0.7, interpolate='natural')
                        ).encode(
    y = alt.Y('Energy:Q', stack='zero',sort=alt.Sort(field='Energy', op='mean', order='ascending')),
    x = alt.X('Year:N', ),

    color=alt.Color("Type:N",),
    order = alt.Order('order:Q', sort='descending')
    # size = alt.Size('Energy:Q')
    # color=alt.Color("Year:N")

    # row = alt.Row('Year:O'),
).transform_joinaggregate(
    order='sum(Energy)',
    groupby=['Type']
)
chart_year_area.properties(
                width=900, height=400
            )
```
